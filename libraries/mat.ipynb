```python
import numpy as np
```

```python
x=np.arange(0,10)
x
```

```python
y=2*x
y
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y)
plt.xlabel('x axis')
plt.ylabel('y axis')
```

```python
plt.plot(x,y)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('straight line')
```

```python
plt.plot(x,y)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('straight line')
plt.show()
```

```python
plt.plot(x,y)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('straight line')
plt.show()
plt.xlim(0,6)
plt.ylim(0,12)
```
